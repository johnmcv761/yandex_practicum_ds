### Проверка расчета эффективности

Проверим правильность расчета recovery на обучающей выборке после этапа флотации (черновой концентрат).

```python
def recovery_calculation(c, f, t):
    return 100 * c * (f - t) / (f * (c - t))
```

```python
train['rougher.output.recovery_check'] = recovery_calculation(c=train['rougher.output.concentrate_au'],
                                                              f=train['rougher.input.feed_au'],
                                                              t=train['rougher.output.tail_au'])

display(train[['rougher.output.recovery_check', 'rougher.output.recovery']].sample(5, random_state=42))

display('MAE между rougher.output.recovery и посчитанными вручную значениями:')
try:
    display(mean_absolute_error(train['rougher.output.recovery'], train['rougher.output.recovery_check']))
except:
    display('Посчитать MAE не удалось')
    display("ValueError: Input contains NaN, infinity or a value too large for dtype('float64').")
```

Визуально видим, что расчеты совпадают, но MAE не можем посчитать из-за наличия "нехороших" значений (NaN, inf) в данных.

```python
try:
    assert len(train['rougher.output.recovery'].dropna()) == len(train['rougher.output.recovery_check'].dropna())
    display('Количество NaN-ов совпадает')
except:
    display('Разница в количестве NaN в исходном датафрейме и рассчитанных вручную значениях')
    display(train['rougher.output.recovery'].isna().sum() -
            train['rougher.output.recovery_check'].isna().sum())
```

Видим, что количество NaN не совпадает: в предоставленном датафрейме их на 343 больше, чем мы получили при ручном расчете. Посмотрим на несовпадающие NaN-ы.

```python
cond1 = (train['rougher.output.recovery'].isna() & train['rougher.output.recovery_check'].notna())
cond2 = (train['rougher.output.recovery'].notna() & train['rougher.output.recovery_check'].isna())

train[cond1 | cond2][['rougher.output.recovery_check', 'rougher.output.recovery']]
```

```python
display('rougher.output.recovery_check')
display(train[cond1 | cond2]['rougher.output.recovery_check'].value_counts(dropna=False))
display('rougher.output.recovery')
display(train[cond1 | cond2]['rougher.output.recovery'].value_counts(dropna=False))
```

Интересно. Во-первых, видим, что нет ситуаций, когда в датафрейме есть значение, а в посчитанных руками данных - NaN. Во-вторых, видим, что среди значений, рассчитанных руками много странных значений: отрицательные значения эффективности, значения, превышащие 100, бесконечности. Посмотрим есть ли среди этих значений значения из ожидаемого интервала значений [0; 100].

```python
cond3 = (0 <= train['rougher.output.recovery_check']) & (train['rougher.output.recovery_check'] <= 100)
train[(cond1 | cond2) & cond3]['rougher.output.recovery_check']
```

Видим, что таких нет. Значит, в `rougher.output.recovery` они как раз отсеяны и помечены как NaN.

Посчитаем MAE, дропнув NaNы в двух рассматриваемых столбцах.

```python
cleaned = train[['rougher.output.recovery_check', 'rougher.output.recovery']].dropna()

display('MAE между rougher.output.recovery и посчитанными вручную значениями:')
mean_absolute_error(cleaned['rougher.output.recovery'], cleaned['rougher.output.recovery_check'])
```

```python
cleaned['delta'] = abs(cleaned['rougher.output.recovery'] - cleaned['rougher.output.recovery_check'])
cleaned[cleaned['delta'] !=0].sort_values(by='delta', ascending=False).head()
```

Видим, что эффективность посчитана корректно, отклонения пренебрежимо малы.

```python
def update_train_test(full=full, train=train, test=test):
    full_dates = full['date']

    display('Размер train до обновления ' + str(len(train)))
    display('Размер train после обновления ' + str(len(train.query('date in @full_dates'))))

    display('Размер test до обновления ' + str(len(test)))
    display('Размер test после обновления ' + str(len(test.query('date in @full_dates'))))

    return train.query('date in @full_dates'), test.query('date in @full_dates')
```

```python
def check_train_test():
    display(f'Доля тестовых данных {len(test)/len(full)}')
    display(f'Доля данных в обучающей выборке {len(train)/len(full)}')
    display(f'Размер full - {len(full)}')
    try:
        assert len(train) + len(test) == len(full)
        display('Сумма размеров train и test равна размеру full')
    except:
        display('Сумма размеров train и test не равна размеру full')
```

```python
train, test = update_train_test(full, train, test)

check_train_test()
```

## Модель

Для решения поставленной задачи попробуем три модели: линейную регрессию, как наиболее легко интерпретируемую,  и случайный лес, который показывал лучшие результаты ранее. Для оценки адекватности результатов будем сравнивать их с предсказаниями по среднему и по медиане.

### Функции для расчета sMape

```python
def sMape(y_true, y_pred):
    assert len(y_true) == len(y_pred)

    n = len(y_true)
    df = pd.DataFrame({'y_true': y_true, 'y_pred': y_pred})

    df['to_sum'] = 2 * abs(df['y_true'] - df['y_pred']) / (abs(df['y_true']) + abs(df['y_pred']))
    result = 100 * df['to_sum'].sum() / n

    return result
```

```python
def final_Mape(y_true, y_pred):
    assert len(y_true) == len(y_pred)

    y_pred = pd.DataFrame(y_pred, columns=['rougher_pred', 'final_pred'])
    y_pred.index = y_true.index

    rougher_true = y_true['rougher.recovery_true']
    final_true = y_true['final.recovery_true']
    rougher_pred = y_pred['rougher_pred']
    final_pred = y_pred['final_pred']

    result = 0.25 * sMape(rougher_true, rougher_pred) + 0.75 * sMape(final_true, final_pred)

    return result
```

```python
final_sMape_scorer = make_scorer(final_Mape)
```

#### Значимость признаков

Посмотрим на признаки, вклад которых оказался наибольшим:

```python
model_lin_regr = LinearRegression()
model_lin_regr.fit(features_train, target_train)

coef_table = pd.DataFrame(model_lin_regr.coef_, index=['rougher_coef', 'final_coef'],
             columns=features_train_no_scaling.columns).transpose()
coef_table['rougher_coef_abs'] = abs(coef_table['rougher_coef'])
coef_table['final_coef_abs'] = abs(coef_table['final_coef'])
display('Признаки, отсортированные по важности для rougher_recovery')
display(coef_table.sort_values(by='rougher_coef_abs', ascending=False).head(15))
display('Признаки, отсортированные по важности для final_recovery')
display(coef_table.sort_values(by='final_coef_abs', ascending=False).head(10))
```

Будем отсекать часть наименее значимых признаков, проварьируем порог, по которому отсекаем и посмотрим, какого лучшего значения финального sMape сможем добиться на кросс-валидации.

```python
max_rougher_coef = coef_table['rougher_coef_abs'].max()
max_final_coef = coef_table['final_coef_abs'].max()

best_threshold = 0
best_cv_lin_regr_mean = cv_results_model_lin_regr.mean()
best_X_col_lin_regr = []

for threshold in tqdm(np.arange(0, 1, 0.05)):
    final_index = coef_table.query('final_coef_abs > @max_final_coef * @threshold').index
    rougher_index = coef_table.query('rougher_coef_abs > @max_rougher_coef * @threshold').index
    X_col_by_coef = set(rougher_index) | set(final_index)
    cv_results_model_lin_regr = cross_val_score(model_lin_regr_cv, features_train[X_col_by_coef], target_train,
                scoring=final_sMape_scorer,
                cv=5)
    result = cv_results_model_lin_regr.mean()
    if result < best_cv_lin_regr_mean:
        best_threshold = threshold
        best_cv_lin_regr_mean = result
        best_X_col_lin_regr = X_col_by_coef

display(f'Лучший средний итоговый sMape = {best_cv_lin_regr_mean} при пороге = {best_threshold}.')
display(f'Названия признаков сохранены в best_X_col_lin_regr. Осталось {len(best_X_col_lin_regr)} признаков')
```

```python
for threshold in tqdm(np.arange(0.2, 0.5, 0.01)):
    final_index = coef_table.query('final_coef_abs > @max_final_coef * @threshold').index
    rougher_index = coef_table.query('rougher_coef_abs > @max_rougher_coef * @threshold').index
    X_col_by_coef = set(rougher_index) | set(final_index)
    cv_results_model_lin_regr = cross_val_score(model_lin_regr_cv, features_train[X_col_by_coef], target_train,
                scoring=final_sMape_scorer,
                cv=5)
    result = cv_results_model_lin_regr.mean()
    if result < best_cv_lin_regr_mean:
        best_threshold = threshold
        best_cv_lin_regr_mean = result
        best_X_col_lin_regr = X_col_by_coef

display(f'Лучший средний итоговый sMape = {best_cv_lin_regr_mean} при пороге = {best_threshold}.')
display(f'Названия признаков сохранены в best_X_col_lin_regr. Осталось {len(best_X_col_lin_regr)} признаков')
```

Видим, что без подбора гиперпараметров и отсева признаков случайный лес дает результаты не лучше, чем DummyRegressor. Посмотрим на значимость признаков

#### Значимость признаков

```python
model_forest = RandomForestRegressor(random_state=42)
model_forest.fit(features_train, target_train)
```

```python
imp_table = pd.DataFrame(model_forest.feature_importances_, index=features_train.columns,
                         columns=['importance']).sort_values(by='importance', ascending=False)
imp_table
```

```python
%%time

model_forest_cv = RandomForestRegressor(n_jobs=-1, random_state=42)

imp_table_index = list(imp_table.index)

deleted = None
best_cv_forest_mean = cv_results_model_forest.mean()
best_X_col_forest = []

for _ in tqdm(range(len(imp_table_index))):
    cv_results_model_forest = cross_val_score(model_forest_cv, features_train[imp_table_index], target_train,
                scoring=final_sMape_scorer,
                cv=5)
    result = cv_results_model_forest.mean()
    if result < best_cv_forest_mean:
        best_cv_forest_mean = result
        best_X_col_forest = imp_table_index.copy()

    deleted = imp_table_index.pop()

display(f'Лучший средний итоговый sMape = {best_cv_forest_mean}')
display(f'Названия признаков сохранены в best_X_col_forest. Осталось {len(best_X_col_forest)} признаков')
display(best_X_col_forest)
```
